**Buatlah 2 variable dengan tipe data integer, lalu buatlah 1 variable yang melakukan perkalian antara 2 variable tersebut.**

```python
def perkalian(x,y):
  return int(x)*int(y) #convert ke int

var_x = input("nilai x adalah :") #secara default fungsi input menginginkan str
var_y = input("nilai y adalah :")

print(perkalian(var_x,var_y))
```

Mencari nilai maksimum dan minimum dengan perulangan

```python
listnum = [4,7,1,9,6,3]

maxnum = listnum[0]
for num in listnum:
  if num > maxnum:
    maxnum = num

print(f"nilai max : {maxnum}")

minnum = listnum[0]
for num in listnum:
  if num < minnum:
    minnum = num

print(f"nilai min : {minnum}")
```

```python
listnum = [4,7,1,9,6,3]

maxnum = max(listnum) #ini function bawaam
print(f"nilai max : {maxnum}")

minnum = min(listnum)
print(f"nilai min : {minnum}")
```

Buatlah sebuah kelas yang mempunyai method parameter wins (3 poin), draws (1 poin) dan losses (0 poin). Lalu hitung jumlah poin sebuah tim sepakbola. 

```python
class timbola():
  def getpoin(self, win, draw, lose): #self menandakan instance method
    self.points = win*3 + draw*1 + lose*0

tim1 = timbola()
tim1.getpoin(20,2,3) #kalo tidak ada nilai balikan maka tidak bisa diset variable
print(tim1.points)     
```

```python
class timbola():
  def getpoin(self, win, draw, lose):
    self.points = win*3 + draw*1 + lose*0
    return self.points

tim1 = timbola()
totalpoin = tim1.getpoin(20,2,3)

print(tim1.points) #atau bisa diprnt
print(totalpoin) 
```

```python
# contoh dengan static method
# jika static method tidak perlu self
# tidak perlu instantiate

class timbola():
  @staticmethod
  def getpoin(win, draw, lose):
    points = win*3 + draw*1 + lose*0
    return points

totalpoin = timbola.getpoin(3,1,6)
print(totalpoin)
```

```python
class timbola():
  @staticmethod
  def getpoin(win, draw, lose):
    points = win*3 + draw*1 + lose*0
    return points

win = input("win : ")
draw = input("draw : ")
lose = input("lose : ")

totalpoin = timbola.getpoin(int(win),int(draw),int(lose))
print(totalpoin)

"""tentang static method
static method hanya bisa ambil class attribut
dan tidak bisa ambil dari method lain.
sedangkan method lain ambil dari static method bisa"""
```

Permainan Rock Paper Scissors

```python
def GBK(p1,p2):
  if p1==p2:
    print("Draw")
  elif p1=="Gunting":
    if p2=="Batu":
      print("p2 menang")
    else:
      print("p2 kalah")
  elif p1=="Batu":
    if p2=="Gunting":
      print("p2 kalah")
    else:
      print("p2 menang")
  elif p1=="Kertas":
    if p2=="Gunting":
      print("p2 menang")
    else:
      print("p2 kalah")
  else:
    print("invalid")

GBK("Batu","Gunting") #karena tidak ada return maka tidak perlu ditampung variable
```

2. filter data yang memiliki "userId" lebih dari 5
3. di setiap datanya tambahkan variabel "jenis_data" dengan value: "genap" 4.jika "id" nya genap, "ganjil" jika "id" nya ganjil
4. cetak data ke 2 dari data yang sudah diproses sebelumnya



```python
import requests

res = requests.get("https://jsonplaceholder.typicode.com/posts")
resjson = res.json()
```

```python
filtered=[]
for i in resjson:
  if i["id"] %2 == 0 :
    i["jenis_data"] = "genap"
  else:
    i["jenis_data"] = "ganjil"
  if i["userId"] >5:
    filtered.append(i)

filtered
```

```python
class Course():
  def __init__(self, title, description, credits):
    self.__title = title #__ menandakan privat, jika privat hanya bisa dipanggil diclass itu aja
    self.__description = description  
    self.__credits = credits
  def getTitle(self): #kalo mau panggil (privat) dari luar maka harus dibuat fungsi
    return self.__title
  def getCredits(self):
    return self.__credits
  def getDescription(self):
    return self.__description

class Instructor():
  def __init__(self, name):
    self.name = name
  def sethomeCampus(self, homeCampus):
    self.homeCampus = homeCampus

class Class():
  def __init__(self, course, code, semester, status, instructor):
    self.course = course
    self.code = code
    self.semester = semester
    self.status = status
    self.instructor = instructor
  
class Student():
  def __init__(self, name):
    self.name = name
  def setEmail(self, email):
    self.email = email
  def setHome(self, home):
    self.home = home
  def setBirthday(self, birthday):
    self.birthday = birthday
  def registerForClass(self, class_obj):
    self.class_obj = class_obj
  
class Undergraduate(Student):
  def __init__(self, name):
    super().__init__(name) #super ini memanggil attribut dari parentnya
    self.parties = []
  def addParty(self, party):
    self.parties.append(party)
  def getParties(self):
    return self.parties

class Graduate(Student):
  def ApplyForAssistance(self, class_obj):
    self.class_obj = class_obj

engineering = Course("engineering","kelas engineer",40)
sutiono = Instructor("sutiono")

sutiono.sethomeCampus("harvard")

mechanic = Class(engineering, "CIIA", "IV", "Aktif", sutiono)

budi = Undergraduate("budi")
budi.setEmail("budi@gmail.com")
budi.setHome("jakarta")
budi.setBirthday("1/3/1999")
#register ke kelas
budi.registerForClass(mechanic)

lusi = Undergraduate("koi")
lusi.setEmail("koi@gmail.com")
lusi.setHome("jepara")
lusi.setBirthday("7/3/1999")
#register ke kelas
lusi.registerForClass(mechanic)

lutfi = Graduate("lutfi")
lutfi.setEmail("lutfi@gmail.com")
lutfi.setHome("jepara")
lutfi.setBirthday("7/9/1999")
lutfi.ApplyForAssistance(mechanic)

print(lusi.class_obj.code)
print(lusi.class_obj.semester)
print(lusi.class_obj.instructor.homeCampus)
print(lusi.class_obj.instructor.name)
```
